```python
import os
import numpy as np
import cv2
import face_recognition
from tqdm import tqdm
import keras
from keras.layers import Activation, Input, Conv2D, Dense, MaxPooling2D, Dropout, Flatten
from keras.models import Sequential, Model
from keras.layers.core import Lambda
from keras.datasets import mnist
from keras import backend as K
from keras.utils import plot_model
import matplotlib.pyplot as plt
import pickle  

# rename emotion files

categories = ["Neutral", "Happy", "Sad", "Surpiced", "Anger", "Disgust", "Fear",  "", "","Kiss"]

images = [[],[],[],[],[],[],[],[],[],[]]
path = "facesdb"
save_path = "faces-jpg100x100"
images_pickle = "images100x100.pickle"
size = 100
learning_rate = 0.00001
input_shape = (size, size, 3)
        
```

```python
def load_images():
    file_format = 'bmp'

    for dir_name in tqdm(os.listdir(path)):
        dir_path = os.path.join(path, dir_name, file_format);
        for file_name in os.listdir(dir_path):
            img = cv2.imread(os.path.join(dir_path, file_name))
            face_location = face_recognition.face_locations(img)[0]
            top, right, bottom, left = face_location
            face_image = img[top:bottom, left:right]
            face_image = cv2.resize(face_image, (size,size))
            label = int(file_name.split('-')[1].split('_')[0])
            images[label].append(face_image)        

#save pickle

def save_pickle():
    file_pickle = open('images{}x{}.pickle'.format(size,size), 'wb') 
    pickle.dump(images, file_pickle)

#load pickle

def load_pickle():
    load_pickle = open(images_pickle, 'rb') 
    return pickle.load(load_pickle)

def save_images():
    for i, emotion in enumerate(images):
        for j, emotion_face in enumerate(emotion):
            file_name = categories[i] + "_" + str(j)
            cv2.imwrite(file_name + ".jpg", emotion_face)

def plot_model(model, name):
    plot_model(model,
            to_file=name,
            show_shapes=True,
            show_layer_names=True,
            expand_nested=True
          )

def siamese_net(input_shape):
    pred_input = Input(input_shape)
    known_input = Input(input_shape)
    model = Sequential()
    model.add(Conv2D(64, kernel_size=(3, 3), name='c1',
                     activation='relu',
                     input_shape=input_shape))
    model.add(Conv2D(64, kernel_size=(3, 3), name='c2', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), name='p1'))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(128, kernel_size=(3, 3), name='c3', activation='relu'))
    model.add(Conv2D(128, kernel_size=(3, 3), name='c4', activation='relu'))
    model.add(Conv2D(256, kernel_size=(3, 3), name='c5', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), name='p2'))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(256, kernel_size=(3, 3), name='c6', activation='relu'))
    model.add(Conv2D(256, kernel_size=(3, 3), name='c7', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), padding='same', name='p3'))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(512, activation='relu', name='d1'))
    model.add(Dense(256, activation='relu', name='d2'))
    encoded_pred = model(pred_input)
    encoded_known = model(known_input)
    L1_layer = Lambda(lambda tensors:K.abs(tensors[0] - tensors[1]))
    L1_distance = L1_layer([encoded_pred, encoded_known])
    prediction = Dense(1,activation='sigmoid',bias_initializer=keras.initializers.he_normal(seed=None))(L1_distance)
    siamese_network = Model([pred_input, known_input], outputs=prediction)
    return siamese_network


def get_max_batches(images_array):
    """
    Create batches of training examples 0.5/0.5 same/different
    """
    pairs = [[],[]]
    targets = []
    number_of_people = 36
    num_classes = images_array.shape[0]
    
    for i in range(num_classes):
        if i == 7 or i == 8:
            continue
        emotion_array = np.array(images_array[i])
        for person_emotion in range(number_of_people):
            random = np.random.randint(number_of_people, size=1)[0]
            while random == person_emotion:
                random = np.random.randint(number_of_people, size=1)[0]
            pairs[0].append(emotion_array[person_emotion, :, : ,:])
            pairs[1].append(emotion_array[random, :, :, :])
            targets.append(1)
    
    print(len(targets))
    
    for i in range(num_classes):
        if i == 7 or i == 8:
            continue
        
        emotion_array = np.array(images_array[i])
        for person_emotion in range(number_of_people):
            random_emotion = np.random.randint(num_classes, size=1)[0]
            while random_emotion == 7 or random_emotion == 8 or random_emotion == i:
                random_emotion = np.random.randint(num_classes, size=1)[0]
            
            random_person = np.random.randint(number_of_people, size=1)[0]
            other_emotion_array = np.array(images_array[random_emotion])
            
            pairs[0].append(emotion_array[person_emotion, :, : ,:])
            pairs[1].append(other_emotion_array[random_person, :, :, :])
            targets.append(0)
            
    pairs2 = np.array(pairs)
    pairs2 = pairs2.reshape((pairs2.shape[0],pairs2.shape[1],pairs2.shape[2],pairs2.shape[3],pairs2.shape[4]))
    pairs = pairs2.tolist()
    return pairs, targets

def get_batch(images_array, batch_size):
    """
    Create max batch of training examples 0.5/0.5 same/different
    """
    batch_size = 70 * 36
    pairs = [[],[]]
    targets = np.zeros((batch_size, 1))
    
    targets[batch_size//2:] = 1
    num_classes = images_array.shape[0]
    for i in range(batch_size//2):
        random = np.random.randint(num_classes, size=1)[0]
        while random == 7 or random == 8:
            random = np.random.randint(num_classes, size=1)[0]
        size = len(images_array[random])
        sample = images_array[random][np.random.randint(size, size=1)[0]]
        random2 = np.random.randint(num_classes, size=1)[0]
        while random2 == random or random2 == 7 or random2 == 8:
            random2 = np.random.randint(num_classes, size=1)[0]
        size2 = len(images_array[random2])
        negative = images_array[random2][np.random.randint(size2, size=1)[0]]
        pairs[0].append(sample)
        pairs[1].append(negative)
    
    
    for i in range(batch_size//2,batch_size):
        random = np.random.randint(num_classes, size=1)[0]
        while random == 7 or random == 8:
            random = np.random.randint(num_classes, size=1)[0]
        size = len(images_array[random])
        sample = images_array[random][np.random.randint(size, size=1)[0]]
        
        size2 = len(images_array[random])
        positive = images_array[random][np.random.randint(size2, size=1)[0]]
        pairs[0].append(sample)
        pairs[1].append(positive)
    
    pairs2 = np.array(pairs)
    pairs2 = pairs2.reshape((pairs2.shape[0],pairs2.shape[1],pairs2.shape[2],pairs2.shape[3],pairs2.shape[4]))
    pairs = pairs2.tolist()
    return pairs, targets

def load_model(path="emotions_weights_500.h5"):
    model.load_weights(path)

def train_siamese_nets(iterations, batch_size = 32, log_rate = 10, save_model_rate = 100):
    for iteration in range(iterations):
        pairs, labels = get_batches(batch_size)
        train_loss, train_accuracy = model.train_on_batch(
            pairs,
            labels)
        if iteration % log_rate == 0:
            print('{}. training loss: {} training accuracy: {}'.format(iteration, train_loss, train_accuracy))
        if iteration % save_model_rate == 0 and iteration > 0:
            model.save_weights('emotions_weights_{}_{}.h5'.format(iterations, batch_size))
```

```python
def find_face_image(image):
    img = cv2.imread(os.path.join(dir_path, file_name))
    face_location = face_recognition.face_locations(img)[0]
    top, right, bottom, left = face_location
    return img[top:bottom, left:right]

def prepare_model_image(image, w=224, h=224):
    face_image = find_face_image(image)
    img = np.array(cv2.resize(face_image, (w,h)))
    img = np.expand_dims(img, 0)
    print(img.shape)
    return img

def place_text_on_image(image, text):
    textsize = cv2.getTextSize(text, font, 3, 2)[0]
    textX = int((image.shape[1] - textsize[0]) / 2)
    textY = int(image.shape[0] - textsize[1])
    cv2.putText(image, text, (textX, textY), font, 3, (0, 0, 255), 2, cv2.LINE_AA)
    
def predict_and_sign(image):
    image2 = prepare_model_image(image)
    pred_class = categories[np.argmax(model.predict(image2))]
    place_text_on_image(image, pred_class)
    return image

def video_detection():
    out2 = cv2.VideoWriter('output_auth.avi', fourcc, 20.0, (640,480))
    cap2 = cv2.VideoCapture('output.avi')         
    for i in tqdm(range(0,100)):
        ret,frame = cap2.read()
        out2.write(predict_and_sign(frame))
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    cap2.release()
    out2.release()
    cv2.destroyAllWindows()
    
def save_video(frames = 100):
    out1 = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480))
    cap1 = cv2.VideoCapture(0) 
    for i in tqdm(range(0, frames)):
        ret,frame = cap1.read()
        out1.write(frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap1.release()
    out1.release()
    cv2.destroyAllWindows()  
```
